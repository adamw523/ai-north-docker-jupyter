```python
import numpy as np
from scipy.cluster import vq
from matplotlib import pyplot as plt
```

## Random convenience function

```python
def rand_floats(minimum, maximum, count):
    return minimum + np.random.sample(count) * (maximum - minimum)
```

```python
x = rand_floats(25, 100, 25)
y = rand_floats(175, 250, 25)
z = np.hstack((x,y))
z = z.reshape((50, 1)) # array size 50, values ranging from 0 to 255
```

```python
p = plt.hist(z, 256, [0,256])
```

```python
centers, dist = vq.kmeans(z, 2)
centers
```

Put the centers of the k clusters on the histogram now

```python
p = plt.hist(z, 256, [0, 256]), plt.hist(centers, 32, [0, 256])
```

#### Adding More Features

```python
w = zip(rand_floats(25, 50, 25), rand_floats(25, 50, 25))  # range 25 to 50, 25 rows, 2 columns
x = zip(rand_floats(60, 85, 25), rand_floats(60, 85, 25))  # range 60 to 85, 25 rows, 2 columns
y = zip(rand_floats(100, 125, 25), rand_floats(100, 125, 25))  # range 100 to 125,25 rows, 2 columns
z = np.vstack((w, x, y))
```

```python
# show the data
plt.scatter(z[:,0], z[:,1], z[:2]); plt.xlabel('Height'); plt.ylabel('Weight')
plt.show()
```

```python
center, dist = vq.kmeans(z, 3)
code, distance = vq.vq(z, center)

a = z[code == 0]
b = z[code == 1]
c = z[code == 2]

plt.scatter(a[:,0], a[:,1]); plt.xlabel('Height'); plt.ylabel('Weight')
plt.scatter(b[:,0], b[:,1], c='r')
plt.scatter(c[:,0], c[:,1], c='g')
plt.scatter(center[:,0], center[:,1], s=80, c='y', marker='s')
plt.show()
```
